```python
#importing libs
import os
import pandas as pd
import logging
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.model_selection import train_test_split
```

```python
#making logs dir with os
logs_dir = 'logs'
os.makedirs(logs_dir,exist_ok=True)
```

```python
#setting logging module
logger  = logging.getLogger('data_ingestion')
logger.setLevel('INFO')

#setting formating 
formatter = logging.Formatter('%(asctime)s - %(name)s-%(levelname)s,%(message)s')

#console handler
console_handler = logging.StreamHandler()
console_handler.setLevel('INFO')
console_handler.setFormatter(formatter)

# setting the file log path
log_path_dir = os.path.join(logs_dir,'data_ingestion_logs')

#file handler
file_handler = logging.FileHandler(log_path_dir)
file_handler.setLevel('INFO')
file_handler.setFormatter(formatter)

# adding looging to console and file handler
logger.addHandler(console_handler)
logger.addHandler(file_handler)
```

```python
# making data loader for the pipeline
def data_load(data_url: str) -> pdDataFrame:
    try:
        df = pd.read_csv(data_url, encoding = 'utf-8')
        logger.info('data loadded sucessfully : %s')
        return df
    except pd.error.Parsererror as e:
        logger.error('failed to parse csv: %s',e)
        raise
    except Exception as e:
        logger.error('unknown error: %s', e)
        raise
```

```python
# preprocessing the data 
def prep_data(df:pd.DataFrame) -> pd.DataFrame:
    try:
        df = df.drop(columns=(['day','year','month','is_weekend','weekday','stock_on_hand','discount_pct','promo_flag','stock_out_flag','lead_time_days','supplier_id','purchase_cost','margin_pct','rain_mm','city','latitude','longitude','sku_name','brand']))
        ohe= OneHotEncoder(handle_unknown='ignore',sparse_output =False).set_output(transform='pandas')
        ohecol = ohe.fit_transform(df[['store_id','sku_id','category','subcategory','country','channel']])
        df = pd.concat([df,ohecol], axis=1).drop(columns=(['store_id','sku_id','category','subcategory','country','channel']))
        logger.info('data preprocessed sucessfully : %s')
        return df
    except KeyError as e:
        logger.error('column not found:%s',e)
        raise
    except Exception as e:
        logger.error('unkkown error has ocuured: %s',e)
        raise    
```
